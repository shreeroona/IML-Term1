Part 2 - Find out association rules using mlxtend aprior and association_rules library

```python
#import library
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#find more frequent artist, had to give min_support.005 else no output was coming with .1 or .01

freq_fm = apriori(df=lfd_binary,min_support=0.01,use_colnames=True)
```

```python
freq_fm.head()
```

```python
rules = association_rules(freq_fm,min_threshold=.5)
```

```python
#Displaying sorted rules in descending order of lift value
rules.sort_values(by='lift',ascending=False)
```
